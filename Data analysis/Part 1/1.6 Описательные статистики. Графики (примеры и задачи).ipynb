```r
library(ggplot2)

df  <- mtcars
df$vs  <- factor(df$vs  , labels = c("V", "S"))
df$am  <- factor(df$am  , labels = c("Auto", "Manual"))
```

```r
hist(df$mpg, breaks = 20, xlab = "MPG")
```

```r
boxplot(mpg ~ am, df, ylab = "MPG")
```

```r
ggplot(df, aes(x=mpg)) + 
    geom_histogram(fill = "white", col = "black", binwidth = 2)
```

```r
ggplot(df, aes(x=mpg)) + geom_dotplot()
```

```r
ggplot(df, aes(x=mpg)) + geom_density(fill = "red", alpha = 0.5)
```

```r
ggplot(df, aes(x=mpg, fill = am)) + geom_dotplot()
```

```r
ggplot(df, aes(x = mpg, fill = am)) + geom_density(alpha = 0.5)
```

```r
ggplot(df, aes(x = am, y = hp)) + 
geom_point()
```

```r
ggplot(df, aes(x = am, y = hp, col = vs)) + 
geom_boxplot()
```

```r
ggplot(df, aes(x = mpg, y = hp)) + 
    geom_point(size = 6)
```

```r
ggplot(df, aes(x = mpg, y = hp, col = vs, size = qsec)) + 
    geom_point()
```
